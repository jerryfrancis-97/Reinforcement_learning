Create environment


```python
import gym
import numpy as np
import random
```

```python
env = gym.make("Taxi-v3")
env.render()
#print(type(env))
```

Create a q-table and initialize it!


```python
action_size = env.action_space.n
print("Action size:", action_size)

state_size = env.observation_space.n
print("State space:", state_size)
```

```python
#q-table

Q_table = np.zeros((state_size,action_size))
print(Q_table)
```

```python
#Hyperparameters

total_episodes = 20000
total_test_episodes = 100
max_steps = 99

learning_rate = 0.7 #alpha
gamma = 0.6 #discount-rate

#exploration paramters
epsilon = 0.5
epsilon_decay = 0.6
max_epsilon = 1.0
min_epsilon = 0.02
```

```python
'''
Q_learning_alogrithm

1. initialise q-table
2. for all life learning
  2.1. choose an action a for the current state based on q-value estimates
  2.2. take action a and observe the next state s' and reward r'
  2.3. update q-value of (s,a) using the SARSAMAX equation
'''

#2. step implemented

for episode in range(total_episodes):
  state = env.reset()
  step = 0
  done = False
  
  for steps in range(max_steps):
    #choosing an action for current state
    exploit_explore_tradeoff = random.uniform(0,1)

    if exploit_explore_tradeoff > epsilon:          #exploitation
      action = np.argmax(Q_table[state,:])          
    
    else:                                           #exploration
      action = env.action_space.sample()
    
    
    # do action a , get next state s' and reward r'
    new_state, reward, done, info = env.step(action)
    
    #update q_table
    Q_table[state][action] = Q_table[state][action] + learning_rate* ( reward + gamma * np.sum(0.25*Q_table[new_state,:])  - Q_table[state, action])
    
    state = new_state
    
    #finishes episode
    if done == True:
      break
  episode += 1
  
  #reduce epsion to move to exploit phase
  
  epsilon = min_epsilon + (max_epsilon - min_epsilon)*np.exp(-epsilon_decay*epsilon)
  
  
    
```

Play taxi!

```python
print(Q_table)
```

```python
import sys
import math

env.reset()

rewards = []
avg_rewards = []
graph = []
best_reward = -math.inf

for episode in range(total_test_episodes+1):
  state = env.reset()
  step = 0
  done = False
  total_rewards = 0
  print("*******************************")
  print("Episode", episode)
  
  for step in range(max_steps):
    env.render()
    action = np.argmax(Q_table[state,:])
    
    new_state, reward, done, info = env.step(action)
    total_rewards += reward
    
    if done:
      rewards.append(total_rewards)
      print("Score:", total_rewards)
      break
    state = new_state
   
  if episode%10 == 0:
    avg_reward = np.mean(rewards)
    avg_rewards.append(avg_reward)
    graph.append((avg_reward, episode))
    print("Aversage reward till now: ", avg_reward)

    if avg_reward > best_reward:
      print("change")
      best_reward = avg_reward

    
  # monitor progress
  
  print("\rEpisode {}/{} || Best average reward {}".format(episode, total_episodes, best_reward), end="")
  sys.stdout.flush()
  # check if task is solved (according to OpenAI Gym)
  if best_reward >= 9.7:
      print('\nEnvironment solved in {} episodes.'.format(episode), end="")
      break
  if episode == total_test_episodes: print('Teest ENded!!!\n')


env.close()
#print("Score over time: " + str(sum(rewards)/total_test_episodes) )
      
```
